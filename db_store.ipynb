```python
# -----------------------------
# STEP 1: Import necessary tools
# -----------------------------
import pandas as pd
import os
from sqlalchemy import create_engine

# -----------------------------
# STEP 2: TiDB Cloud credentials
# -----------------------------
username = "3ZpzE69BSNXuvY2.root"
password = "MSiewQGRtx2F7qYu"
host = "gateway01.us-west-2.prod.aws.tidbcloud.com"
port = 4000
database = "stock_analysis"  # You must create this database manually in TiDB Console
ssl_certificate_path = "certs/ca-certificate.crt"

# -----------------------------
# STEP 3: Build SQLAlchemy connection URL with SSL
# -----------------------------
connection_url = (
    f"mysql+mysqlconnector://{username}:{password}@{host}:{port}/{database}"
    f"?ssl_ca={ssl_certificate_path}"
)

engine = create_engine(connection_url)

# -----------------------------
# STEP 4: Upload CSV files
# -----------------------------
csv_folder = "csv_data"  # Your folder containing stock CSVs

# First run: replace the table to ensure clean structure
first_file = True

for file_name in os.listdir(csv_folder):
    if file_name.endswith(".csv"):
        symbol = file_name.replace(".csv", "")
        file_path = os.path.join(csv_folder, file_name)

        try:
            df = pd.read_csv(file_path)

            # Normalize column names
            df.columns = [col.lower() for col in df.columns]

            # Add the symbol column
            df['symbol'] = symbol

            # Upload to TiDB
            if_exists_mode = 'replace' if first_file else 'append'
            df.to_sql('stock_data', con=engine, if_exists=if_exists_mode, index=False)

            print(f"Uploaded data for '{symbol}' successfully.")
            first_file = False  # only replace on the first file

        except Exception as e:
            print(f"Could not upload data for '{symbol}': {e}")
```
